```python

# Step 1: Import Libraries

import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
```

```python

# Step 2: Data Ingestion

df = pd.read_csv("Reviews.csv", nrows=50000)

print("First 50,000 rows loaded successfully.")
print("Total Rows:", len(df))
```

```python

# Step 3: Text Selection

light_reviews = df[df["Text"].str.contains(r"\blight\b", case=False, na=False)]

print(light_reviews["Text"].head(20))
```

```python

# Step 4: Create List of Three Sentences

sentences = [
    "This olive oil is wonderful and very light on the stomach.",
    "The packaging was broken and let too much light inside the box.",
    "I prefer this brand because it is a light and healthy snack."
]
```

```python

# Step 5: Vectorization Engine

vectorizer = CountVectorizer()

document_term_matrix = vectorizer.fit_transform(sentences)

dense_matrix = document_term_matrix.toarray()

print(dense_matrix)
```

```python

# Step 6: Vector Slicing

vec_1 = dense_matrix[0]
vec_2 = dense_matrix[1]
vec_3 = dense_matrix[2]

print(vec_1)
print(vec_2)
print(vec_3)
```

```python

# Step 7: Similarity Mathematics

similarity_1_2 = np.dot(vec_1, vec_2)

similarity_1_3 = np.dot(vec_1, vec_3)
```

```python

# Step 8: Console Output

print("Sentence 1 vs Sentence 2 (Different Contexts)")
print("Similarity Score:", similarity_1_2)

print("Sentence 1 vs Sentence 3 (Similar Contexts)")
print("Similarity Score:", similarity_1_3)
```
