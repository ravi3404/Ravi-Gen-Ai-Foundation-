```python
import google.generativeai as genai

genai.configure(api_key="AQ.Ab8RN6LZhDXJjlRhzBK76mmUXGWOMN4fz1pKpuzoN1XVDXwIZA")

model = genai.GenerativeModel("gemini-2.5-flash")
```

```python
domain = input("Enter your domain: ")

prompt = f'''
Generate exactly 5 interview questions for {domain}.

Rules:
- Number them 1 to 5
- Questions only
- No answers
'''

response = model.generate_content(prompt)

questions = response.text

print("\nINTERVIEW QUESTIONS\n")
print(questions)
```

```python
answers = []

for i in range(1, 6):
    ans = input(f"Answer {i}: ")
    answers.append(ans)
```

```python
evaluation_prompt = f'''
You are an expert interviewer.

Domain:
{domain}

Questions:
{questions}

Candidate Answers:
{answers}

Generate a detailed report:

1. Overall Score (/100)
2. Technical Score (/100)
3. Communication Score (/100)
4. Strengths
5. Weaknesses
6. Missing Concepts
7. Learning Roadmap
8. Hiring Recommendation
'''

report = model.generate_content(evaluation_prompt)

print("\n" + "="*60)
print("INTERVIEW REPORT")
print("="*60)
print(report.text)
```
